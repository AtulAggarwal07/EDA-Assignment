- Ans.

Step 1: Calculate Q1 (25th percentile) and Q3 (75th percentile) for km_driven.

Step 2: Compute IQR = Q3 - Q1.

Step 3: Determine lower bound = Q1 - 1.5 * IQR and upper bound = Q3 + 1.5 * IQR.

Step 4: Remove entries where km_driven is below the lower bound or above the upper bound.
